```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm

!gdown 1UpLnYA48Vy_lGUMMLG-uQE1gf_Je12Lh
```

```python
df = pd.read_csv('cars24-car-price-clean.csv')
df.head()
```

```python
y=df[['selling_price']]
X=df.drop('selling_price', axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
y_train = np.array(y_train)
```

```python
X_sm = sm.add_constant(X_train) ## Statmodels default is without intercept, to add intercept we need to add constant

model = sm.OLS(y_train, X_sm)
results = model.fit()

print(results.summary())
```

```python
# VIF
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif = pd.DataFrame()
#X_t = pd.DataFrame(X_train, columns=X_train.columns)
vif['Features'] = X_train.columns
vif['VIF'] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
vif = pd.DataFrame()
#X_t = pd.DataFrame(X_train, columns=X_train.columns)
X_train_new = X_train.drop("age", axis=1)
vif['Features'] = X_train_new.columns
vif['VIF'] = [variance_inflation_factor(X_train_new.values, i) for i in range(X_train_new.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
X_sm = sm.add_constant(X_train_new) ## Statmodels default is without intercept, to add intercept we need to add constant

model = sm.OLS(y_train, X_sm)
results = model.fit()

print(results.summary())
```
